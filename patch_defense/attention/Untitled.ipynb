```python
import numpy as np
import cv2
import pickle as pl

import matplotlib.pyplot as plt

a1=np.load("vgg3_freeze_featuremap3_attack_relu.npy")
print(a1.shape)
a1 = np.mean(a1, axis=0)
print(a1.shape)
# a1=cv2.resize(np.mean(a1, axis=0),(32,32),interpolation=cv2.INTER_CUBIC)
a1=cv2.resize(a1,(224,224))
plt.matshow((a1))
plt.colorbar()

print(np.mean(a1))
print(a1[1][1])


 
b = np.zeros(63*63)
print(b)
k = 0
for i in range(39, 102):
    for j in range(12, 75):
        b[k] += a1[i][j]
        k += 1
print("average value of patch is:", np.mean(b))
print("max value of patch is:", np.max(b))


ar = a1.copy()
for i in range(39,102):
    for j in range(12, 75):
        ar[i][j] = 0
        
print("average value of non_patch is:", np.sum(ar)/(224*224 - 63*63))
print("max value of non_patch is:", np.max(ar))
print(np.where(ar == np.amax(ar)))
```
